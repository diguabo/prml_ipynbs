```python
from sklearn.datasets import load_digits
digits = load_digits()
print "images shape: {0}, targets shape: {1}".format(digits.images.shape, digits.target.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.matshow(digits.images[0], cmap=plt.cm.Greys)
```

```python
digits.target  # 各サンプルが何の数字であるかを表すベクトル(正解データ)
```

```python
# Prepare the data
X = digits.data.reshape(-1, 64)  # 8x8だった画像データを64次元のベクトルにreshape => (サンプル数, 64次元ベクトル)
print X.shape
y = digits.target
print y.shape
```

```python
# reshapeの動作の確認
import numpy as np

test_mat = np.array([[[1,2], [3,4]],
                     [[5,6], [7,8]]])  # 2x2x2
print "2x2x2"
print test_mat
print "2x4"
print test_mat.reshape(-1, 4)  # -1でベクトルの1次元目の要素の順番が保たれる
```

```python
# scikit-learnにおける重要な原則: 入力データは常に(サンプル数, 特徴量数)のshape
print X
```

```python
# PCAによる次元削減
from sklearn.decomposition import PCA

pca = PCA(n_components=2)  # 基底数を2に設定
pca.fit(X)  # 入力データから基底を計算
X_pca = pca.transform(X)  # 入力データを計算した基底に対し射影し次元を削減
X_pca.shape
```

```python
plt.figure(figsize=(16,10))  # pdfのplt.figsizeは動作しない
# 2次元に削減された全サンプルを散布図プロット、色の指定をc=yとすることで番号ごとに色を変える
plt.scatter(X_pca[:,0], X_pca[:,1], c=y)
plt.xlabel("pca feature 1")
plt.ylabel("pca feature 2")
```

```python
# legendを付けるには数字ごとにプロットする必要がある
# 参考 https://stackoverflow.com/questions/26558816/matplotlib-scatter-plot-with-legend
# colormapについてはこちらを参考にした
# http://seesaawiki.jp/met-python/d/matplotlib
# http://d.hatena.ne.jp/bettamodoki/20110628/1309242049
import matplotlib.cm as cm

plt.figure(figsize=(16,10))  # pdfのplt.figsizeは動作しない

for i in range(10):
    xs = X_pca[:, 0][y == i]
    ys = X_pca[:, 1][y == i]
    plt.scatter(xs, ys, c=cm.jet(float(i)/10.0), label=i)
    
plt.xlabel("pca feature 1")
plt.ylabel("pca feature 2")
plt.legend()
```

上記の結果から、たとえばfeature1が0付近、feature2が-25付近であればその数字が0であると区別できそうなことが分かる。

```python
print pca.mean_.shape
print pca.components_.shape

fig, axs = plt.subplots(1, 3)
axs[0].matshow(pca.mean_.reshape(8,8), cmap=plt.cm.Greys)
axs[1].matshow(pca.components_[0, :].reshape(8,8), cmap=plt.cm.Greys)
axs[2].matshow(pca.components_[1, :].reshape(8,8), cmap=plt.cm.Greys)
```

PCAの基底は元データを中心化した上で固有値分解をした際の固有ベクトルであるため、0を合成するためには以下のようにする

```python
zero_f1_mean = np.mean(X_pca[:, 0][y == 0])
zero_f2_mean = np.mean(X_pca[:, 1][y == 0])

plt.matshow(zero_f1_mean * pca.components_[0, :].reshape(8,8)
            + zero_f2_mean * pca.components_[1, :].reshape(8,8)
            + pca.mean_.reshape(8,8),
            cmap=plt.cm.Greys)
```

なお、これと同等のことを行うためのinverse_transform()も用意されている

```python
zero_origin = pca.inverse_transform([[zero_f1_mean, zero_f2_mean]])
plt.matshow(zero_origin.reshape(8,8), cmap=plt.cm.Greys)
```

PCAで2次元に縮約した場合、復元できる数字とできない数字がある気配があるが、これはPCAが線形変換であることの限界を表しているものと思われる。
そこで非線形変換である多様体学習を試してみる。
多様体学習については以下が分かりやすい。
http://www.slideshare.net/kohta/risomap

```python
from sklearn.manifold import Isomap

# 多様体学習, n_neighborsについては上記リンク参照
isomap = Isomap(n_components=2, n_neighbors=20)
isomap.fit(X)
X_isomap = isomap.transform(X)
print X_isomap.shape

# plot
plt.figure(figsize=(16,10))  # pdfのplt.figsizeは動作しない

for i in range(10):
    xs = X_isomap[:, 0][y == i]
    ys = X_isomap[:, 1][y == i]
    plt.scatter(xs, ys, c=cm.jet(float(i)/10.0), label=i)
    
plt.xlabel("isomap feature 1")
plt.ylabel("isomap feature 2")
plt.legend()
```

PCAよりも識別しやすくなっているように見える。

PDFにはないがついでにLDA(線形判別分析)も試す。

```python
from sklearn.lda import LDA

# 多様体学習, n_neighborsについては上記リンク参照
lda = LDA(n_components=2)
lda.fit(X, y)   # LDAはtargetの値も与える必要がある
X_lda = lda.transform(X)
print X_lda.shape

# plot
plt.figure(figsize=(16,10))  # pdfのplt.figsizeは動作しない

for i in range(10):
    xs = X_lda[:, 0][y == i]
    ys = X_lda[:, 1][y == i]
    plt.scatter(xs, ys, c=cm.jet(float(i)/10.0), label=i)
    
plt.xlabel("lda feature 1")
plt.ylabel("lda feature 2")
plt.legend()
```

0, 4, 6は区別しやすいが他は依然混ざり合っている

## classification ##

ここからは話題が変わりクラス分類の話題となる。
最初にアルゴリズムの評価のために学習セットと評価セットにデータセットを分割する。

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

print "X_train shape: {0}".format(repr(X_train.shape))
print "y_train shape: {0}".format(repr(y_train.shape))
print "X_train shape: {0}".format(repr(X_test.shape))
print "y_train shape: {0}".format(repr(y_test.shape))
```

train_test_splitはデフォルトだと3/4と1/4の量にデータを分割する。これらのサイズは引数のtest_sizeまたはtrain_sizeで指定可能である。また引数のrandom_stateにはintかRandomStateのインスタンスを指定可能。詳しくはipython上でsklearn.cross_validation.train_test_split?を実行して確認すること。

## 線形カーネルSVMによる分類 ##

```python
from sklearn.svm import LinearSVC

svm = LinearSVC()
svm.fit(X_train, y_train)
result = svm.predict(X_train)  # 学習データと同じデータで予測

print result   ## 予測結果
print y_train  ## 正解
```

```python
# 学習データに対しモデルを評価
svm.score(X_train, y_train)
```

```python
# 評価データに対しモデルを評価
svm.score(X_test, y_test)
```

## RandomForestによる分類 ##

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf.score(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

SVMよりも良いスコアとなっている

## モデル選択と評価 ##

評価データを最後まで分けて取っておくために、学習データ内をn等分して、(n-1)/nのデータを学習データ、1/nのデータを評価データとし、これをn回繰り返すクロスバリデーションを試す。これにより学習データへのオーバーフィッティングを防ぐ。

```python
from sklearn.cross_validation import cross_val_score

scores = cross_val_score(rf, X_train, y_train, cv=5)  # n=5のクロスバリデーション
print "scores: {0} mean: {1} std: {2}".format(str(scores), np.mean(scores), np.std(scores))
```

```python
# より沢山の決定木を使うRandomForestを学習する(デフォルトは10)
rf2 = RandomForestClassifier(n_estimators=50)
scores = cross_val_score(rf2, X_train, y_train, cv=5)  # n=5のクロスバリデーション
print "scores: {0} mean: {1} std: {2}".format(str(scores), np.mean(scores), np.std(scores))
```

性能が上がっていることが分かる

### 重要なパラメータをグリッドサーチで調整する ##

パラメータを順番に振ってモデルが一番良い性能となるパラメータを探ることをグリッドサーチという。

ここでは線形SVMにおいて誤りに対するペナルティパラメータであるCを、クロスバリデーションしながらグリッドサーチを行うsklearn.grid_search.GridSearchCVで調整する。なおパラメータCは、値が大きくなるほど誤差を許さなくなるためハードマージンに近づき、Cが小さいとある程度の誤差を許容した上で識別境界を学習するソフトマージンとなる。
参考 http://yut.hatenablog.com/entry/20121009/1349739006

```python
from sklearn.grid_search import GridSearchCV

param_grid = {"C": 10. ** np.arange(-5, 4)}

# scikit-learn 0.15.2ではcompute_training_scoreオプションは廃止されている
# http://scikit-learn.org/stable/modules/generated/sklearn.grid_search.GridSearchCV.html#sklearn.grid_search.GridSearchCV
grid_search = GridSearchCV(svm, param_grid=param_grid, cv=3, verbose=3)  # cvはクロスバリデーションのn
grid_search.fit(X_train, y_train)
```

```python
print grid_search.best_params_, grid_search.best_score_
print ""
print grid_search.grid_scores_
print ""
print [c.mean_validation_score for c in grid_search.grid_scores_]
print [c.cv_validation_scores for c in grid_search.grid_scores_]
print [c.parameters for c in grid_search.grid_scores_]
```

```python
# 現在のバージョンはPDFとはgrid_searchの仕様が大きく変わっている模様
plt.figure(figsize=(12, 6))
plt.plot([c.mean_validation_score for c in grid_search.grid_scores_], label="validation score")
plt.xticks(np.arange(6), param_grid["C"])
plt.xlabel("C")
plt.ylabel("Accuracy")
plt.legend(loc="best")
```

PDFでは学習データにおけるスコアとバリデーション用データにおけるスコアの比較をしているが、0.15.2では学習データのスコアが取り出せないためここでは比較できない。
一般にPDFにおける図の右側のように学習データに対しては性能が良いがテストデータに対しては駄目な場合をオーバーフィッティング(バリアンスが大きい)、逆にPDFにおける図の左側のように学習データに対してもテストデータに対しても性能がでない場合をアンダーフィッティングと呼んでいる(バイアスが大きい)。
前者の場合、求められた平均は一般的なものではない。後者の場合、平均は学習データにおいてさえも間違っている。

## SNSにおける侮辱的コメントの検出 ##

フォーラムへの短い投稿が侮辱的なものかを分類する。
コメントの分類は単純なbag of wordsによる分析でも効果が高い(bag of wordsとは単語の出現順は無視して頻度情報だけ用いる手法)。なおテキストデータに対しては線形識別器がたいていベストな性能が出る。
使用するデータは以下から取得する。
https://github.com/amueller/sklearn_tutorial

```python
import pandas as pd

train_data = pd.read_csv("kaggle_insult/train.csv")
test_data = pd.read_csv("kaggle_insult/test_with_solutions.csv")
```

"Insult"カラムが侮辱的コメントか否かを表し、"Comment"カラムがテキストデータを含む

```python
y_train = np.array(train_data.Insult)
comments_train = np.array(train_data.Comment)
print comments_train.shape
print y_train.shape
```

```python
print comments_train[0]
print "Insult: {0}".format(y_train[0])
```

```python
print comments_train[5]
print "Insult: {0}".format(y_train[5])
```

これらのコメントをsklearn.feature_extraction.textモジュールのCountVectorizerを使ってbag of wordsを抽出する。
sklearn.feature_extractionモジュールはテキストや画像などのデータから特徴量を抽出するためのモジュールである。

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
cv.fit(comments_train)  # ここで辞書を学習する

print cv.get_feature_names()[:15]
```

```python
print cv.get_feature_names()[1000:1015]
```

```python
print len(cv.get_feature_names())  # 単語数
```

```python
# transform(): 単語-文書行列のsparse matrixを返す
# [[0, 3, 4.,..,2],   # 0番目の文書の語彙に対する頻度
   [3, 4, 2,...,0]]]  # 1番目の...
# tocsr():  sparse matrixをCSR形式に変換、CSRは掛け算や足し算が速く、行を取り出すのが速い疎行列の形式
# 参考 http://docs.scipy.org/doc/scipy-0.14.0/reference/sparse.html
X_train = cv.transform(comments_train).tocsr()
print "X_train.shape: {0}".format(X_train.shape)

print X_train[0, :]  # 単語文書行列における0番目の文書の単語頻度ベクトルを表す
print cv.get_feature_names()[3409]

print X_train[0:5, :]  # 0, 1, 2, 3, 4目の文書の単語頻度ベクトル
```

## 線形SVM分類器を学習する ##

```python
from sklearn.svm import LinearSVC

svm = LinearSVC()
svm.fit(X_train, y_train)
```

```python
comments_test = np.array(test_data.Comment)
y_test = np.array(test_data.Insult)
X_test = cv.transform(comments_test)
svm.score(X_test, y_test)
```

```python
# 上手く行った例
print comments_test[8]
print "Target: {0}, prediction: {1}".format(y_test[8], svm.predict(X_test.tocsr()[8])[0])
```
